# Feature Enginnering

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, scale
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import SMOTENC, SMOTE
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("Data/data.csv")
df
```

<font color= "green">First it will remove the variables that are not relevant to the machine learning model.</font> <font color= "blue"> And include numeric variable for target variable to optimize the model. 

```python
df2 = df
df2["PROFILE_N"] = df2.PROFILE.map({"EXCELLENT": 5,
                   "VERY_GOOD": 4,
                   "HUMAN": 3,
                   "EXACTS": 2,
                   "DIFFICULTY": 1})
df2
```

```python
columns= ["DISAPPROVALS_BL","DISAPPROVALS_EN","DISAPPROVALS_FM","DISAPPROVALS_OM","NOTE_BL","NOTE_EN","NOTE_FM",
          "NOTE_OM","ENGLISH","H_CLASS_PRES","ONLINE_TASKS","ABSENCES","PROFILE_N"]
df2 = df2[columns]
```

```python
df2
```

treating missing values

```python
# Instantiating imputation algorithm
imputation = SimpleImputer(missing_values= np.nan, strategy="median")
```

```python
# imputing missing values
df2 = pd.DataFrame(imputation.fit_transform(df2), columns= columns)
```

```python
df2
```

```python
df2.isnull().sum()
```

```python
df2.duplicated().sum()
```

```python
df2.drop_duplicates(inplace= True)
```

```python
df2.duplicated().sum()
```

Generating new variables

```python
bins = [-1, 2, 4, 6, 8, 10]
var = ["NOTE_BL", "NOTE_EN", "NOTE_FM", "NOTE_OM"]
```

```python
a = 0
for i in var:
    df2[a] = pd.cut(df2[i], bins= bins)
    a+=1
df2.rename(columns={0:"NOTE_BL_CAT", 1:"NOTE_EN_CAT", 2:"NOTE_FM_CAT", 3:"NOTE_OM_CAT"}, inplace=True)    
```

```python
df2.head()
```

```python
df2.shape
```

```python
df2.dtypes
```

```python
df2.isna().sum()
```

```python
df2.dropna(inplace=True)
```

```python
df2.duplicated().sum()
```

```python
df2 = df2.reindex(columns= ["DISAPPROVALS_BL","DISAPPROVALS_EN","DISAPPROVALS_FM","DISAPPROVALS_OM","NOTE_BL",
                            "NOTE_EN","NOTE_FM","NOTE_OM","ENGLISH","H_CLASS_PRES","ONLINE_TASKS","ABSENCES",
                            "NOTE_BL_CAT", "NOTE_EN_CAT", "NOTE_FM_CAT", "NOTE_OM_CAT","PROFILE_N"])
df2
```

```python
df2.shape
```

```python
# Saving dataframe
df2.to_csv("Data/data2.csv", index= False)
```

<font color="red">Generating training and testing data</font> 

```python
X = df2.iloc[:, 0:16]
Y = df2.PROFILE_N
x_train, x_test, y_train, y_test= train_test_split(X, Y, test_size= .20, random_state= 102)
```

```python
x_train.dtypes
```

```python
x_train.shape
```

<font color="green">Training data balancing</font>

```python
# Instantisting balancing algorithm
balanced = SMOTENC(categorical_features=[12,13,14,15] ,random_state= 42)
```

```python
x_train.dropna(inplace=True)
x_train, y_train = balanced.fit_resample(x_train, y_train)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
# Saving Training and testing dataframe
x_train.to_csv("Data/x_train.csv", index= False)
x_test.to_csv("Data/x_test.csv", index= False)
y_train.to_csv("Data/y_train.csv", index= False)
y_test.to_csv("Data/y_test.csv", index= False)
```

<font color="green">Standardizing data</font>

```python
column = ["DISAPPROVALS_BL","DISAPPROVALS_EN","DISAPPROVALS_FM","DISAPPROVALS_OM","NOTE_BL",
          "NOTE_EN","NOTE_FM","NOTE_OM","ENGLISH","H_CLASS_PRES","ONLINE_TASKS","ABSENCES"]
```

```python
x_train_norm= x_train.iloc[:, 0:12]
```

```python
scale= StandardScaler()
x_train_norm = pd.DataFrame(scale.fit_transform(x_train_norm), columns= column)
x_train_norm
```

```python
x_test_norm= x_test.iloc[:, 0:12]
x_test_norm= pd.DataFrame(scale.transform(x_test_norm), columns= column)
x_test_norm
```

```python
# Saving standardized training dataframe
x_train_norm.to_csv("Data/x_train_norm.csv", index= False)
x_test_norm.to_csv("Data/x_test_norm.csv", index= False)
```
